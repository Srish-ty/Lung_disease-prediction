```python
import numpy as np
import pandas as pd
import os
import librosa
#from tqdm.notebook import tqdm
# "" 
path = "./Respiratory_Sound_Database/audio_and_txt_files" 
files = []
for f in os.listdir(path):
    if f.split(".")[-1]=="wav":
        files.append(f)
#files
```

```python
x = np.array(files)
#print(x)
print(x.shape)
```

```python
patient_dia= pd.read_csv("./Respiratory_Sound_Database/patient_diagnosis.csv")
patient_dia.set_index('pid',inplace=True)
patient_dia
```

```python
diseases = patient_dia['disease']
diseases
```

```python
diseases.value_counts()
```

```python
y_labels=[]

for file in files:
    pid = int(file.split("_")[0])
    y_labels.append(diseases[pid])

y_labels[:10]
```

```python
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()
y= label_encoder.fit_transform(y_labels)
y[:20]
```

```python
files[:10]
```

```python
#sr =22050
file0, fs = librosa.load(path+"/"+files[0])
file0.shape
```

```python
file0
```

```python
mfcc0= librosa.feature.mfcc(y=file0, sr=fs)
mfcc0.shape
```

```python
mfcc0
```

```python
mfcc0_data=pd.DataFrame(mfcc0)
mfcc0_data
```

```python
X=[]
```

```python
for file in (files):
    fl, sr = librosa.load(path+"/"+file)
    mfccs = librosa.feature.mfcc(y=fl)
    X.append(np.mean(mfccs, axis=1))  
```

```python
X=np.array(X)
X.shape
```

```python
X = pd.DataFrame(X)
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, shuffle=True)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()

model.fit(X_train, y_train)
```

```python
y_actu = label_encoder.inverse_transform(y_test)
y_pre = model.predict(X_test)
y_hat = label_encoder.inverse_transform(y_pre)
```

```python
comp ={'Predicted values':y_hat, 'Actual values':y_actu}
pd.DataFrame(comp)
```

```python
accu=model.score(X_test, y_test)
accu
```

```python
print("The Accuracy is",accu*100 ,"%")
```

```python
y_pred_prob= model.predict_proba(X_test)
y_pred_prob
```

```python
pred_wid_prob = []

for pred in y_pred_prob:
    stri=""
    for i in pred:
        if i>0.1:
            j = np.where(pred==i)[0]
            dise = label_encoder.inverse_transform(j)
            stri+=( str(round(i*100,2))+"% "+dise[0]+", ")
        #stri+="\n"
    pred_wid_prob.append(stri)
pred_wid_prob
```

```python
final_res ={ 'Given Disease':y_hat ,'Our prediction':pred_wid_prob }
results = pd.DataFrame(final_res)
results
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_actu, y_hat)
```
