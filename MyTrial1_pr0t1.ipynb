```python
import numpy as np
import os
import librosa
 
path = "C:/Users/srush/Downloads/Stethaim/my_Trial1" 
files = os.listdir(path+"/COPD")
x = np.array(files)
print(x)
print(x.shape)
```

```python
import pandas as pd

chosen_pids= pd.read_csv(path+"/ylabels/chosen_pid.csv")
diseases = chosen_pids['Disease']
diseases
```

```python
diseases.value_counts()
```

```python
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()
y= label_encoder.fit_transform(diseases)
y
```

```python
#sr =22050
file0, fs = librosa.load(path+"/COPD/"+files[0])
file0.shape
```

```python
file0
```

```python
mfcc0= librosa.feature.mfcc(y=file0, sr=fs)
mfcc0.shape
```

```python
mfcc0
```

```python
mfcc0_data=pd.DataFrame(mfcc0)
mfcc0_data
```

```python
X=[]
```

```python
for file in files:
    fl, sr = librosa.load(path+"/COPD/"+file)
    mfccs = librosa.feature.mfcc(y=fl)
    X.append(np.mean(mfccs, axis=1))
```

```python
X=np.array(X)
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()

model.fit(X_train, y_train)
```

```python
y_actu = label_encoder.inverse_transform(y_test)
y_pre = model.predict(X_test)
y_hat = label_encoder.inverse_transform(y_pre)
```

```python
comp ={'Predicted values':y_hat, 'Actual values':y_actu}
pd.DataFrame(comp)
```

```python
model.score(X_test, y_test)
```
